```python
import re
from nltk.tokenize import sent_tokenize
from collections import Counter

def remove_special_chars(text):
    """Loại bỏ ký tự đặc biệt và chuẩn hóa khoảng trắng"""
    # Giữ lại chữ tiếng Việt và dấu câu cơ bản
    text = re.sub(r'[^\s\w\.,!?\-áàảãạăắằẳẵặâấầẩẫậéèẻẽẹêếềểễệóòỏõọôốồổỗộơớờởỡợíìỉĩịúùủũụưứừửữựýỳỷỹỵđ]', ' ', text)
    return re.sub(r'\s+', ' ', text).strip()

# Apply the remove_special_chars function to the text_content column
df['text_content'] = df['text_content'].apply(remove_special_chars)

df['text_content'] = df['text_content'].str.replace('\n', ' ')
# Tokenize the text_content into sentences
df['sentences'] = df['text_content'].apply(sent_tokenize)

# Flatten the list of sentences and count the occurrences of each sentence
all_sentences = [sentence for sentences in df['sentences'] for sentence in sentences]
sentence_counts = Counter(all_sentences)
ordered_sentence_counts = sentence_counts.most_common()
print('ordered_sentence_counts', ordered_sentence_counts[:10])
```

```python
# Define a threshold for considering a sentence as repetitive
threshold = 15

# Filter out repetitive sentences
df['filtered_sentences'] = df['sentences'].apply(lambda sentences: [sentence for sentence in sentences if sentence_counts[sentence] < threshold])

# Join the filtered sentences back into a single string
df['text_content'] = df['filtered_sentences'].apply(' '.join)

# Drop the intermediate columns
df.drop(columns=['sentences', 'filtered_sentences'], inplace=True)

# Export the cleaned dataframe to a new CSV file
df.to_csv('cleaned_link_spider_results.csv', index=False)
```

```python
# Read the cleaned CSV file
cleaned_df = pd.read_csv('cleaned_link_spider_results.csv')

# Display the first few rows of the cleaned dataframe
print("First few rows of the cleaned dataframe:")
print(cleaned_df.head())

# Display basic information about the cleaned dataframe
print("\nBasic information about the cleaned dataframe:")
print(cleaned_df.info())

# Display summary statistics of the cleaned dataframe
print("\nSummary statistics of the cleaned dataframe:")
print(cleaned_df.describe())
```
